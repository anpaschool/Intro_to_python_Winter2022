```python
# Relational plots
## Numbers
```

```python
sns.relplot(x='Age', y= 'Fare', data = df, kind = 'scatter', hue = 'Sex')
```

```python
sns.relplot(x='Age', y= 'Fare', data = df, kind = 'scatter', hue = 'Sex', col='Pclass')
```

```python
sns.relplot(x='Age', y= 'Fare', data = df, kind = 'scatter', hue = 'Sex', row='Pclass')
```

```python
sns.relplot(x='Age', y= 'Fare', data = df, kind = 'scatter', hue = 'Sex', aspect = 2, style = 'Pclass')
```

```python
sns.relplot(x='Age', y = 'Fare', data= df, kind ='line', ci = None,hue = 'Pclass', palette= ['b','r','g'])
```

```python
## Catgorical Plots
# https://seaborn.pydata.org/generated/seaborn.catplot.html
```

```python
sns.catplot(y='Sex', kind = 'count', data =df, hue = 'Survived' )
```

```python
sns.catplot(x='Sex', y ='Age',
            kind ='bar', hue = 'Survived',
            data =df, ci = None,
            order = ['female','male'])
```

```python
sns.catplot(x='Pclass',  y= 'Age', kind = 'point' , data  =df, hue ='Sex', join = False, aspect = 1.5, ci = None)
```

```python
## Seaborn has "style"
#sns.set_style('whitegrid')
sns.catplot(x='Pclass',  y= 'Age', kind = 'point' , data  =df, hue ='Sex', join = False, aspect = 1.5, ci = None)
```

```python
sns.catplot(x='Pclass',  y= 'Age', kind = 'point' ,
            data  =df, hue ='Sex', join = False,
            aspect = 1.5, ci = None, palette='Purples')
```

```python

sns.set_palette("RdBu")
sns.catplot(x='Pclass', y = 'Age', kind = 'bar', data =df)
plt.savefig('class2.png')
```
